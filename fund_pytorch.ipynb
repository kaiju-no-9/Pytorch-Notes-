```python
# tensor shape issue !!
torch_A = torch.tensor([[1,2],
                       [2,3],
                      [3,4]])
torch_B= torch.tensor([[3,4], [4,5],[6,7]])
```

```python
# Swtiching and transpose
print(f"origal_shape : {torch_A.shape} , {torch_B.shape}")
print("......")
print(f"new  matrix : {torch_A.shape} , {torch_B.T.shape}")
print("......")
print(torch.matmul(torch_A , torch_B.T))

```
