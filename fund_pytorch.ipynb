 .reshapping refers to reshaping tensors

 .stacking refers to.      .stacking one tensor over the other

 .permute- return view of the input with dimention


```python
x= torch.arange(1,10)
x, x.shape
```

```python
# reshape works in the format of factor so for s matrix of
# 10 we can do (5,2) or (2,5) as the reult isthe multiple is 10
x_reshaped= x.reshape(1,10)
x_reshaped
```

```python
#same as reshape the diffrence is that it shares
# shape from the orignal input
z = x.view(2,5)
z, z.shape
```

```python
#Stack function in torch
x_stacked = torch.stack([x,x,x] , dim =0)
x_stacked
```

```python
print(f" prev :{x} ")
print(f"newShape: {x_reshaped} , {x_reshaped.shape}")
x_squeeeze = x_reshaped.squeeze()
print(f"squeezed: {x_squeeeze} , {x_squeeeze.shape}")
```

```python
# addes a single tensor at a specific dim
print(f"previous targert : {x_squeeeze} ")
x_unsqueezed= x_squeeeze.unsqueeze(dim=0)
print(f"new targert : {x_unsqueezed} , {x_unsqueezed.shape} ")
```

```python
# trouch.permute. rerrange the value of tensor in specific order
x_original= torch.rand(size=(224,224,5))
# we have tenssor dimmintion for [height, width, colour_channels]

# permute for origanal tensor to rearracvnge the axis (or dim ) order
x_permuted= x_original.permute(2,0,1)
print(f"previous:{x_original.shape}")
print(f"new:{x_permuted.shape}")
```

```python
# Repoducibility(used with a seed to reduce randomness)
random_tensor_A= torch.rand(3,4)
random_tensor_B = torch.rand(3,4)
print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
random_seed =82
# rand seed
torch.manual_seed(random_seed)
random_tensor_C= torch.rand(3,4)
# rand seed again
torch.manual_seed(random_seed)
random_tensor_D= torch.rand(3,4)
print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D )
```
